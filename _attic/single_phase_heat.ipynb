```python


import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt


cp = 12020
area_side = 0.01/250
area_cross = 0.01**2
htc = 0.0104202
cond = 0.0174
T_amb = 300
J = 0*12e-3

A = -J*cp/cond
B = htc*area_side/(cond*area_cross)
C = -htc*area_side*T_amb/(cond*area_cross)

# Define the system of ODEs
def system(u, x):
    u1, u2 = u
    du1dx = u2
    du2dx = -A*u2 - B*u1 - C
    return [du1dx, du2dx]

# Initial conditions
u0 = [20, 0]  # Assuming y(0) = 0 and y'(0) = 0

# Time grid
x = np.linspace(0, 0.01, 250)

# Solve the system
sol = odeint(system, u0, x)

# Plot the solution
plt.plot(x, sol[:, 0], label='T(x)')
# plt.plot(x, sol[:, 1], label="T'(x)")
plt.legend()
plt.xlabel('t')
plt.ylabel('T [K]')
plt.title('Solution of the second-order ODE')
plt.grid(True)
plt.show()

```
